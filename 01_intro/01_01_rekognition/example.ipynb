```python
import boto3
import base64

rekognition = boto3.client(
    'rekognition',
    #aws_access_key_id=,
    #aws_secret_access_key=,
    region_name='ap-northeast-2'
)

# with : 파일이 열고 닫히는 것을 자동으로 관리해주는 역할
# => 파일작업중에 예외가 발생해도 안전하게 닫아준다.
# rb : binary모드로 읽겠다는 의미
with open ('./images/pikachu.jfif', 'rb') as cf:
    
    # AWS Rekognition으로 전송하기 위해 바이너리 데이터로 전처리
    base64_image = base64.b64encode(cf.read())
    base64_binary = base64.decodebytes(base64_image)
    print("file open!")


response = rekognition.detect_labels(Image={'Bytes': base64_binary})

response
```
